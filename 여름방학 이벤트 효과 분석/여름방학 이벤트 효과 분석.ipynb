## 캐릭터 생성 데이터에서 필요한 컬럼 가져오기

```python
df_new = df[['characterId', 'characterName', 'jobName', 'jobGrowName', 'date']]
df_new = df_new.copy()
df_new.rename(columns = {'date' : '캐릭터 생성 날짜'}, inplace = True)
df_new
```

## 최고 레벨 달성 데이터에서 필요한 컬럼 가져오기

```python
df1_new = df1[['characterId', 'characterName', 'jobName', 'jobGrowName', 'date']]
df1_new = df1_new.copy()
df1_new.rename(columns = {'date' : '최고 레벨 달성 날짜'}, inplace = True)
df1_new
```

## 두 데이터 합치기

```python
total_df = pd.merge(df_new, df1_new, how = 'left', left_on = 'characterId', right_on = 'characterId')
total_df_new = total_df[['characterId', 'characterName_x', '캐릭터 생성 날짜','최고 레벨 달성 날짜', 'jobName_x', 'jobGrowName_x']]
total_df_new = total_df_new.copy()
total_df_new.rename(columns = {'characterName_x' : '캐릭터명', 'jobName_x' : '전직명', 'jobGrowName_x' : '직업명'}, inplace = True)
total_df_new
```

## 날짜 데이터 변환하기

```python
total_df_new['캐릭터 생성 날짜'] = total_df_new['캐릭터 생성 날짜'].str.split(" ").str.get(0)
total_df_new['최고 레벨 달성 날짜'] = total_df_new['최고 레벨 달성 날짜'].str.split(" ").str.get(0)
total_df_new
```

## 필요한 지표 추가

### 이벤트 전/후

```python
total_df_new1 = total_df_new.copy()
total_df_new1['캐릭터 생성 날짜'] = pd.to_datetime(total_df_new1['캐릭터 생성 날짜'])
total_df_new1['전/후'] = np.where(total_df_new1['캐릭터 생성 날짜'].values >= pd.to_datetime('2023-06-29'), '후', '전')
total_df_new1
```

### 성장 속도

```python
total_df_new1['최고 레벨 달성 날짜'] = pd.to_datetime(total_df_new1['최고 레벨 달성 날짜'])
total_df_new1['성장속도'] = total_df_new1['최고 레벨 달성 날짜'] - total_df_new1['캐릭터 생성 날짜']
total_df_new1['성장속도'] = total_df_new1['성장속도'].dt.days + 1
total_df_new1
```

### 하루 만에 만렙 성공 여부

```python
total_df_new1['성공여부'] = np.where(total_df_new1['성장속도'].values == 1, '성공', '실패')
total_df_new1
```

## 결측값 처리

* 캐릭터 생성 날짜와 최고 레벨 달성 날짜가 모두 결측치가 없는 경우
    * 2023-04-21 ~ 2023-07-20 사이에 캐릭터 생성부터 최고 레벨 달성까지 한 경우 -> 성장 속도도 정확하게 구할 수 있고 하루 만에 최고레벨을 달성했는지 여부도 판별 가능
* 캐릭터 생성 날짜만 결측치인 경우
    * 캐릭터 생성은 3개월 전에 했지만 최근에 최고 레벨을 달성했다 -> 하루만에 최고 레벨 달성 x
* 최고 레벨 달성 날짜만 결측치인 경우
    * 아직 최고 레벨을 달성 못했거나 게임을 더 이상 플레이하지 않는 캐릭터 -> 하루 만에 최고 레벨 달성 x
* 둘 다 결측치인 경우
    * 캐릭터 생성, 최고 레벨 달성 모두 3개월 전에 했다. -> 알 수 없음
